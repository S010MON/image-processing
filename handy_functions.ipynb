```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.cvtColor(cv2.imread('project2/images/face01_o.jpeg'), cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Zero Padding
padded = cv2.copyMakeBorder(img, 100, 100, 100, 100, cv2.BORDER_CONSTANT)
plt.imshow(padded)
```

```python
# Rotation

def rotate(img, theta):
    # get the the center of the image
    (h, w) = img.shape[:2]
    (cX, cY) = (w // 2, h // 2)

    # rotate the image by theta degrees around the center of the image
    M = cv2.getRotationMatrix2D((cX, cY), theta, 1.0)
    rotated = cv2.warpAffine(img, M, (w, h))
    
    return rotated
```

```python
plt.imshow(rotate(img, 37))
```
